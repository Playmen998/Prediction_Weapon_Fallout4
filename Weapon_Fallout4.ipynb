Создадим датафрайм split_bomb для последующей работы 

```python
"Функция проверяет есть ли вхождения из списка damage"
def allpresent(text, damage):
    for damag in damage:
        while text.find(damag) != -1:
            return True
    return False
```

```python
"Функция фильтрует данные урона"
def damage_sorting(tabel):
    columns = ['damage','electrical','radiation','frost','fire','poison',"blood","explosion"]
    df = pd.DataFrame(columns = columns)
    damage = columns[1:]

    for column in columns:
        df[column] = tabel
        
        if column.lower() not in damage:
            df[column] = df[column].str.split('+').apply(lambda x: x[0] if (len(x) == 2) or (len(x) == 3) else x)
            df = df.applymap(lambda x: x[0] if isinstance(x, list) else x)
            df[column] = df[column].apply(lambda x: x.replace(f"{x}",'0') if allpresent(x, damage) else x) 
    
        else:
            df[column] = df[column].apply(lambda x: x.replace(f"{x}",'0') if column.lower() not in x else x) #
            df[column] = df[column].str.split('+').apply(lambda x: x[1] if (len(x) == 2) else x)
            df = df.applymap(lambda x: x[0] if isinstance(x, list) else x)
            df[column] = df[column].apply(lambda x: x.replace(f' ({column.lower()})', ''))
    return df
```

```python
split_bomb = damage_sorting(bomb["Урон"])
```

В столбце "УронСекунду" возможные пропуски из-за того, что оружие имеет один выстрел. Согласно статье [fallout.fandom.com](https://fallout.fandom.com/ru/wiki/%D0%A3%D1%80%D0%BE%D0%BD_%D0%B2_%D1%81%D0%B5%D0%BA%D1%83%D0%BD%D0%B4%D1%83)

Получается для этого оружия выстрел происходит слишком долго (больше секунды). Заполним данные ппропуски 0, т.к. за 1 секунду оружие не успевает нанести урон

## Подсчет кол-во патронов для полного поражения цели

Построим модель предсказания **кол-во патронов** для убийства существа. Посчитаем это значение с помощью формул, а далее будем предсказывать это значение для тестовой выборки.  
Чтобы посчитать этот параметр отбросим большую часть столбцов в таблице, будем использовать только те, которые *непосредственно* и *наверняка* участвуют в подсчете параметра (кол-во патронов)  
В расчете формул будем непосредственно отталкиваться от информации наносимого урона в игре fallout 4

**Damage Resistance**  
Посмотрим как существа получают урон на основе характеристик сопротивления к урону. Для этого обратимся к статье на сайте [Fandom](https://fallout.fandom.com/wiki/Damage_Resistance#Fallout_4) в которой отлично расписана формула коэффициента нанесения урона от оружия. 

$$ DamageCoeff =  0.99 * \left(\frac{fPhysicalDamageFactor * Damage}{DamageResist} \right)^{0.365} $$
+ fPhysicalDamageFactor - коэффициент урона, в игре по умолчанию 0.15  
+ Damage - наносимый урон оружием
+ DamageResist - параметр сопротивлению к урону
+ DamageCoeff - процент наносимого урона от атаки оружием  
Данный коэффициент умножается на урон от оружия, это и будет наносимый урон здоровью существа

Для комбинированного типа повреждения (когда оружие может наносить как физический, так и радиционный урон и т.д.) одно попадание пули расценивается, как комбинированный урон от разных типов повреждения.

Для такого урона как radiation, fire, frost, blood не будем высчитывать дополнительные временные эффекты урона.  
Для вида урона fire, frost, blood, explosion нет параметров сопративления у существ, поэтому этот тип урона будем высчитывать как обычный физический

**Range affect damage**

Параметр "Дальность" говорит нам о расстоянии, после которого урон патрона снижается. Согласно статье на [Fandom](https://fallout.fandom.com/wiki/Range#Fallout_4):
> максимальное расстояние, на котором конкретное оружие будет наносить максимальный урон за пределами VATS и дополнительно влияет на шанс поразить цель при использовании VATS

Согласно странице сопротивления урону в [Fandom](https://fallout.fandom.com/wiki/Damage_Resistance#Fallout_4) , урон от оружия за пределами этого максимального диапазона уменьшит урон до 50% при удвоении диапазона
>RangeMultiсоставляет 1,0 при 100% дальности оружия или ближе, 0,5 при 200% WeaponRangeили более, и масштабируется от 1,0 до 0,5 по мере увеличения дальности от 100% до 200% от WeaponRange.

Таким образом, для облегчения расчетов будем считать, что расстояния выстрела всегда оптимальное и не переходит черту, тем самым при каждом выстреле будет наносится 100% урон

**Accuracy Fallout 4**

Точность в Fallout 4 влияет на разброс пуль согласно ответу одному пользователю на форуме [gamespot.com](https://gamefaqs.gamespot.com/boards/164594-fallout-4/72871814)
>Точность в большинстве игр типа fps относятся к разбросу выстрела после его выстрела, а также к разбросу выстрела от бедра большинство оружия с низкой точностью будет иметь огромное поле разброса, что означает, что даже если ваша прицельная сетка находится в центре врага, ваша пуля все еще может оказаться далеко слева от прицельной сетки. именно так точность используется в этой игре.

Из-за того, что стрельба ведется с оптимального расстояния, то процент промахов будет достаточно низким, поэтому не будем брать во внимания данную метрику

остальные метрики, мы не будем высчитывать т.к. 
1. Нет информации и формул, как эти показатели связаны друг с другом, например характеристика "Восприятия" у существ
2. Другие переменные трудно интепретировать, например "Агрессия", "Помощь" и т.д. данные характеристики влияют на контекст встречи с существом, поэтому отбросим их
3. Ненужные данные, например тип патронов и т.д.

```python
columns = ['Название_существа', 'Здоровье','Сопротивление_ФизУрон', 'Сопротивление_ЭнергоУрон', 'Сопротивление_РадУрон',
           'Сопротивление_ЯдУрону', 'Название', 'Тип_Оружия', 'damage', 'electrical', 'radiation', 'frost', 'fire', 
           'poison', 'blood', 'explosion']
```

```python
"""
Функция считает кол-во патронов необходимое для убиства существа
"""
def ammo_count(df,x):
    map_damage = {'damage' : 'Сопротивление_ФизУрон',
                 'fire' : 'Сопротивление_ФизУрон',
                 'frost' : 'Сопротивление_ФизУрон',
                 'blood' : 'Сопротивление_ФизУрон',
                 'explosion' : 'Сопротивление_ФизУрон',
                 'electrical' : 'Сопротивление_ЭнергоУрон',
                 'radiation' : 'Сопротивление_РадУрон',
                 'poison' : 'Сопротивление_ЯдУрону'}
    
    name_columns = list(map_damage.keys())
    
    dmg = []
    for column in name_columns:
        fPhysicalDamage = 0.15
        damage = df[column].loc[x] #номер строки
        DamageResist = df[map_damage[column]].loc[x]
        dmg.append(df[column].loc[x] * (((fPhysicalDamage * damage) / DamageResist) ** 0.365))
    dmg_one_ammo = np.sum(dmg)
    hp = df['Здоровье'].loc[x]
    ammo = 0
    while hp > 0:
        hp -= dmg_one_ammo
        ammo += 1
    return ammo
```

```python
from hyperopt import hp
```

```python
ctb_reg_params = {
    'learning_rate':     hp.loguniform('learning_rate', -5, 0),
    'l2_leaf_reg':       hp.choice('l2_leaf_reg', np.arange(1, 8, 1)),
    'max_depth': hp.quniform('max_depth', 2, 10, 1),
    'n_estimators':  hp.quniform('n_estimators',  1500, 5000, 500),
    'random_seed' : 42,
    'eval_metric': hp.choice('eval_metric', ['RMSE','MAE','MAPE']),
    
    'cat_features': categorical_features_indices
}
ctb_fit_params = {
    'early_stopping_rounds': 100,
    'verbose': False
}
ctb_para = dict()
ctb_para['reg_params'] = ctb_reg_params
ctb_para['fit_params'] = ctb_fit_params
ctb_para['loss_func' ] = lambda y, pred: mean_absolute_percentage_error(y, pred) #здесь мы указываем функцию потерь
```

```python
from hyperopt import fmin, tpe, STATUS_OK, STATUS_FAIL, Trials
```

```python
x_train, x_val, y_train, y_val = train_test_split(df[X],df[y],train_size=0.8,random_state=42)
```

```python
class HPOpt(object):

    def __init__(self, x_train, x_val, y_train, y_val):
        self.x_train = x_train
        self.x_val  = x_val
        self.y_train = y_train
        self.y_val  = y_val

    def process(self, fn_name, space, trials, algo, max_evals):
        fn = getattr(self, fn_name)
        try:
            result = fmin(fn=fn, space=space, algo=algo, max_evals=max_evals, trials=trials)
        except Exception as e:
            return {'status': STATUS_FAIL,
                    'exception': str(e)}
        return result, trials
    
    def ctb_reg(self, para):
        reg = CatBoostRegressor(**para['reg_params'])
        return self.train_reg(reg, para)

    def train_reg(self, reg, para):
        reg.fit(self.x_train, self.y_train,
                eval_set=[(self.x_train, self.y_train), (self.x_val, self.y_val)],**para['fit_params'])
        pred = reg.predict(self.x_val)
        loss = para['loss_func'](self.y_val, pred)
        return {'loss': loss, 'status': STATUS_OK}
```

```python
obj = HPOpt(x_train, x_val, y_train, y_val)

ctb_opt = obj.process(fn_name='ctb_reg', space=ctb_para, trials=Trials(), algo=tpe.suggest, max_evals=100)
```
